```python
import pandas as pd

# Carregar o conjunto de dados

df = pd.read_csv('/content/basedeDados.csv')
# Visualizar as primeiras linhas do conjunto de dados
print(df.head())

# Informações básicas sobre o conjunto de dados
print(df.info())

```

```python
# Verificar valores ausentes
print(df.isnull().sum())
```

```python
#LIMPEZA
#remover linhas com valores ausentes
df = df.dropna()

# Verificar os tipos de dados
print(df.dtypes)
```

```python
# Estatísticas descritivas
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt # Import matplotlib for plotting

#Calcular a matriz de correlação
correlation_matrix = df.drop('Country', axis=1).corr() # Exclude the 'Country' column

# Visualizar a matriz de correlação
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()

# Check if 'Life Satisfaction' is in the DataFrame
if 'Life Satisfaction' in df.columns:
    # Correlações com a Satisfação com a Vida
    correlations = correlation_matrix['Life Satisfaction'].sort_values(ascending=False)
    print(correlations)
else:
    print("Warning: 'Life Satisfaction' column not found in the DataFrame.")
    print("Available columns:", df.columns)
```
